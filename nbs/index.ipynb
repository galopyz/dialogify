```sh
# make sure dialogify package is installed in development mode
$ pip install -e .

```

## Usage

### Installation

Install latest from the GitHub [repository][repo]:

```sh
$ pip install git+https://github.com/galopyz/dialogify.git
```

or from [conda][conda]

```sh
$ conda install -c galopyz dialogify
```

or from [pypi][pypi]


```sh
$ pip install dialogify
```
